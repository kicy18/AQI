```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df['PM 2.5'] = df['PM 2.5'].apply(lambda x : round(x , 2))
df
```

```python
df.drop(['Unnamed: 0'] , axis = 'columns' , inplace = True)
```

```python
df.describe()
```

```python
def zscores( dfcolumn):
    zscore = (df[dfcolumn] - df[dfcolumn].mean())/df[dfcolumn].std()
    return zscore
```

```python
df['zscore_t'] = zscores('T')
```

```python
df
```

```python
df.zscore_t[(df.zscore_t >3) & (df.zscore_t <-3)]
```

```python
df.drop(['zscore_t'] , axis = 'columns' , inplace = True)
```

```python
df['zscore_tm'] = zscores('TM')
df.zscore_tm[(df.zscore_tm >3) & (df.zscore_tm <-3)]
```

```python
df.drop(['zscore_tm'] , axis = 'columns' , inplace = True)
```

```python
X = df.drop(['PM 2.5'] , axis = 'columns')
Y = df['PM 2.5']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 10)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(X)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Lasso
from sklearn.svm import SVR
```

```python


model_param = {
    'svc': {
        'model': SVR(gamma='scale'),
        'params': {
            'C': [0.1, 1, 10],
            'kernel': ['linear'],
            'epsilon': [0.1]
        }
    },
    'random_regressor': {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': [100, 50],
            'random_state': [30 , 43],
          
            'min_samples_split': [2]
        }
    },
    'gradient_boosting': {
        'model': GradientBoostingRegressor(),
        'params': {
            'n_estimators': [100 , 50],
            'max_depth': [3],
            'min_samples_split': [2],
            'learning_rate': [0.01]
        }
    },
    'lasso': {
        'model': Lasso(),
        'params': {
            'alpha': [0.01, 0.1],
            'max_iter': [1000],
            'tol': [0.01]
        }
    },
    'linear': {
        'model': LinearRegression(),
        'params': {}
    }
}


```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
import pandas as pd

# Assuming X and Y are defined
scores = []
for model_name, mp in model_param.items():
    clf = RandomizedSearchCV(mp['model'], mp['params'], n_iter=10, cv=5, n_jobs=-1)
    clf.fit(x_train_scaled, Y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

score_df = pd.DataFrame(scores, columns=('model', 'best_score', 'best_params'))
print(score_df)
```

```python
from sklearn.ensemble import StackingRegressor
base_models = [
    ('rf', RandomForestRegressor(n_estimators=100)),
    ('gb', GradientBoostingRegressor(n_estimators=100)),
    ('svr', SVR(gamma='scale'))
]
meta_model = LinearRegression()
stacked_model = StackingRegressor(estimators = base_models, final_estimator=meta_model)
```

```python
stacked_model.fit(x_train_scaled , Y)
```

```python
stacked_model.score(x_test_scaled , y_test)
```

```python
stacked_model_randomforest = StackingRegressor(estimators = base_models, final_estimator=RandomForestRegressor())
stacked_model_randomforest.fit(x_train_scaled , Y)
stacked_model_randomforest.score(x_test_scaled , y_test)
```
